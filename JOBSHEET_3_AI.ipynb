# **LATIHAN**

**1.	Central Tendency**

**a.	Mendapatkan Nilai Mean**

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
print(x)
```

**b.	Mendapatkan Nilai Median**

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
print(x)
```

**c.	Mendapatkan Nilai Modus / Mode**

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score, keepdims=True)
print(x)
```

**2.	Spread**

**a.	Mendapatkan Nilai Range**

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum
print(x)
```

**b.	Mendapatkan Nilai Percentile**

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.percentile(data, 90)
print(x)
```

**c.	Mendapatkan Nilai Quartile dan Interquartile Range**

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)
```

**d.	Mendapatkan Nilai Variance dan Standart Deviation**

```python
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)
print(mean)
print(var)
print(std)
```

```python
from scipy import stats

# Data pelemparan dadu
data_dadu = [1, 2, 3, 4, 5, 5, 6, 2, 3, 4, 5, 6, 6, 4, 3, 2, 1, 4, 3, 5, 6, 6, 5, 4, 5]

# Menghitung modus
hasil_modus = stats.mode(data_dadu, keepdims=True)

print("--- JAWABAN NOMOR 1 ---")
print(f"Modus: {hasil_modus.mode[0]}")
print(f"Frekuensi: {hasil_modus.count[0]} kali muncul")
```

```python
import numpy as np

# Data deret
data_deret = [9, 10, 11, 6, 8, 7, 7, 5, 4, 5]

# Menghitung persentil
p25 = np.percentile(data_deret, 25)
p75 = np.percentile(data_deret, 75)

print("--- JAWABAN NOMOR 2 ---")
print(f"Persentil ke-25: {p25}")
print(f"Persentil ke-75: {p75}")
```

**3.**

```python
import numpy as np

# Data hasil ujian
data_ujian = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# a. Mengurutkan data
data_urut = sorted(data_ujian)

# b. Q1, Q3, dan IQR
q1 = np.percentile(data_ujian, 25)
q3 = np.percentile(data_ujian, 75)
iqr = q3 - q1

# c. Lower Bound & Upper Bound
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# d. Identifikasi Outlier
outliers = [x for x in data_ujian if x < lower_bound or x > upper_bound]

print("--- JAWABAN NOMOR 3 ---")
print(f"Data Terurut: {data_urut}")
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")
print(f"Batas Bawah: {lower_bound}, Batas Atas: {upper_bound}")
print(f"Data Outlier: {outliers}")
```

```python
import pandas as pd
import numpy as np

# Input data tabel
data = {
    'Berat': ['41-45', '46-50', '51-55', '56-60', '61-65', '66-70'],
    'f': [2, 3, 6, 4, 5, 2],
    'xi': [43, 48, 53, 58, 63, 68] # Nilai tengah (misal: (41+45)/2 = 43)
}

df = pd.DataFrame(data)

# Langkah hitung Standar Deviasi
n = df['f'].sum()
df['f_xi'] = df['f'] * df['xi']
mean = df['f_xi'].sum() / n

df['deviasi_kuadrat'] = (df['xi'] - mean)**2
df['f_deviasi_kuadrat'] = df['f'] * df['deviasi_kuadrat']

varians = df['f_deviasi_kuadrat'].sum() / (n - 1)
std_deviasi = np.sqrt(varians)

print("--- JAWABAN NOMOR 4 ---")
print(f"Rata-rata (Mean): {mean:.2f}")
print(f"Standar Deviasi: {std_deviasi:.2f}")
```
